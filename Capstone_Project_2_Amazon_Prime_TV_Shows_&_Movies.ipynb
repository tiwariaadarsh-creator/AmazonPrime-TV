## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# lets drop that columns that have null values > 50%.

titles.info()
````

````python
titles.drop(columns = ['age_certification','seasons'],inplace=True)
````

````python
titles.columns    # columns---> age_certification, seasons has been removed.
````

````python
titles.info()
````

````python
# lets replace null values which is present in columns---> description, imdb_id,imdb_score, imdb_votes, tmdb_popularity and tmdb_score.

titles['description'] = titles['description'].fillna('-9999')
titles['imdb_id'] = titles['imdb_id'].fillna('-9999')
titles['imdb_votes'] = titles['imdb_votes'].fillna('-9999')
titles['tmdb_popularity'] = titles['tmdb_popularity'].fillna('-9999')
titles['tmdb_score'] = titles['tmdb_score'].fillna('-9999')
titles['imdb_score'] = titles['imdb_score'].fillna('-9999')
````

````python
titles.isnull().sum()       # here all null values changed to -9999. That's why no null values are shown in our dataframe.
````

````python
credits.info()   # here column---> character have null values in it.
````

````python
# lets replace these null values with -9999.

credits['character'] = credits['character'].fillna('-9999')
````

````python
credits.isnull().sum()   # all null values of character column changed to -9999.
````

````python
# convert 'id' column of titles and credits dataframe to string.

titles['id'] = titles['id'].astype(str)
credits['id'] = credits['id'].astype(str)
````
